```python
import pandas as pd
df = pd.read_csv('drive/MyDrive/Datasets_1/NDCLD15-with-Segmentation.csv')
```

```python
df.head()
```

```python
k = df.drop(columns = ['date',	'format'	,'subjectid',	'eye'	,'color','gender'	,'tag_list'	,'pupilX'	,'pupilY'	,'pupilR'	,'irisX',	'irisY',	'irisR'],axis = 1)
k.head()
def label_race(row):
  if row['contacts_texture']=='Yes':
    return 'texture'
  elif row['contacts_cosmetic']=='Yes':
    return 'texture'
  elif row['contacts'] =='Yes':
    return 'soft'
  else:
    return 'nolens'

import numpy as np
# k = k.drop[df['contacts_texture'] =='NaN']
k.drop(k.index[k['contacts_texture'] == 'Unknown'], inplace=True)
k.drop(k.index[k['contacts_texture'] == 'NaN'], inplace=True)
k.shape
k.dropna(subset=['contacts_texture'],inplace=True)
k.drop(k.index[k['contacts_cosmetic'] == 'Unknown'],inplace = True)
k.dropna(subset=['contacts_cosmetic'],inplace=True)
k['cls'] = k.apply (lambda row: label_race(row), axis=1)
k.head()
k['cls'].value_counts()
k = k.drop(columns = ['contacts','contacts_type','contacts_texture','contacts_toric','contacts_cosmetic'],axis = 1)
k.head()
from sklearn.model_selection import train_test_split
train,test = train_test_split(k, test_size=0.1780, random_state=0)
strArr = np.empty(5874, dtype='object')
strArr_1 = np.empty(1273,dtype='object')
for i in range(0,5874):
  strArr[i] = 'train'
for i in range(0,1273):
  strArr_1[i] = 'test'
train['train'] = strArr
test['test'] = strArr_1


```

```python
#define function to swap columns
def swap_columns(df, col1, col2):
    col_list = list(df.columns)
    x, y = col_list.index(col1), col_list.index(col2)
    print(x,y)
    temp = col_list[x]
    col_list[x] = col_list[y]
    col_list[y] = temp
    # col_list[y], col_list[x] = col_list[x], col_list[y]
    print(col_list)
    df = df[col_list]
    return df
# train = swap_columns(train, 'sequenceid', 'train')
col_list = ['train','sequenceid','cls']
col_list_1 = ['test','sequenceid','cls']
train = train[col_list]
test = test[col_list_1]
# train_1 = train
# test
train['sequenceid'] = train['sequenceid'] + '.tiff'



```
